Write a program to implement 
1) Vanilla Gradient Descent            
2) Momentum based gradient descent 

Consider 2 points : (2,0.3) (6,0.8) Approximate these 2 points by using above mentioned algorithm, also plot the approximation curve. 

Learning Rate : 0.5

```python
# Vanilla Gradient Descent
import numpy as np 

X = [2,0.3]
Y= [6,0.8]
xpoints=[]
ypoints=[]


def f(x,w,b):
    return 1/(1+np.exp(-(w*x+b)))

def error(w,b):
    err=0.0
    for x,y in zip(X,Y):
        fx=f(x,w,b)
        err +=(fx-y)**2
    return 0.5*err

def grad_b(x,w,b,y):
    fx=f(x,w,b)
    return (fx-y)*fx*(1-fx)

def grad_w(x,w,b,y):
    fx=f(x,w,b)
    return (fx-y)*fx*(1-fx)*x


def do_gradient_descent():
    
    xpoints=[]
    ypoints=[]
    w,b,eta,max_epoches = 0.359,0,0.5,100
    
    
    for i in range(max_epoches):
        dw,db=0,0
        for x,y in zip(X,Y):
            dw +=grad_w(x,w,b,y)
            db +=grad_b(x,w,b,y)
            
        w=w-eta*dw
        xpoints.append(w)
        b=b-eta*db
        ypoints.append(b)
        
    return xpoints, ypoints
    
        

        
```

```python
import matplotlib.pyplot as plt



xpoints, ypoints = do_gradient_descent()
for i in range(len(xpoints)):
    print(xpoints[i],ypoints[i])

plt.plot(xpoints, ypoints)
# naming the x axis
plt.xlabel('Weights')
# naming the y axis
plt.ylabel('Bais')
  
plt.show()
```

```python
# Momemtum Based Gradient Descent 
import numpy as np
X = [2,0.3]
Y= [6,0.8]
mgd_xpoints=[]
mgd_ypoints=[]

def f(x,w,b):
    return 1/(1+np.exp(-(w*x+b)))

def grad_b(x,w,b,y):
    fx=f(x,w,b)
    return (fx-y)*fx*(1-fx)

def grad_w(x,w,b,y):
    fx=f(x,w,b)
    return (fx-y)*fx*(1-fx)*x

def do_mgd(max_epochs):
    w,b,eta = -2,-2,1.0
    prev_uw,prev_ub,beta = 0,0,0.9
   
    for i in range(max_epochs):
        dw,db = 0,0        
        for x,y in zip(X,Y):
            dw += grad_w(w,b,x,y)
            db += grad_b(w,b,x,y)
            
        uw = beta*prev_uw+eta*dw
        ub = beta*prev_ub+eta*db
        w = w - uw
        mgd_xpoints.append(w)
        b = b - ub
        mgd_ypoints.append(b)
        prev_uw = uw
        prev_ub = ub
    return mgd_xpoints, mgd_ypoints

        
```

```python
import matplotlib.pyplot as plt



xpoints, ypoints = do_mgd(40)
print(xpoints)
plt.plot(xpoints, ypoints)
# naming the x axis
plt.xlabel('Weights')
# naming the y axis
plt.ylabel('Bais')
  
plt.show()
```
